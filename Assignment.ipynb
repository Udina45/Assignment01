```python

import numpy as np
arr = np.random.randint(1, 101, (5, 5))
print("Original Array:\n", arr)


even_mask = arr % 2 == 0
arr[even_mask] = np.sqrt(arr[even_mask])
print("Modified Array:\n", arr)

x = np.linspace(-3, 3, 50)
sigmoid = 1 / (1 + np.exp(-x))
print("Sigmoid:\n", sigmoid)


matrix = np.random.randint(1, 100, (4, 4))
print("Matrix:\n", matrix)
print("2nd Column:\n", matrix[:, 1])
print("Diagonal:\n", np.diag(matrix))
print("Odd Numbers:\n", matrix[matrix % 2 != 0])
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Carol', 'Dave', 'Eva', 'Frank', 'Grace'],
    'Age': [23, 25, 22, 24, 21, 26, 23],
    'Gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'Marks': [85, 58, 90, 45, 76, 67, 92],
    'Department': ['CSE', 'ECE', 'CSE', 'ME', 'ECE', 'CSE', 'CSE']
}

df = pd.DataFrame(data)
print(df.shape)
print(df.info())
print(df.describe())


df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 60 else 'Fail')


def grade(m):
    if m >= 90: return 'A'
    elif m >= 80: return 'B'
    elif m >= 70: return 'C'
    elif m >= 60: return 'D'
    else: return 'F'
df['Grade'] = df['Marks'].apply(grade)


print(df.groupby('Department')['Marks'].mean())
print(df['Department'].value_counts())


print(df[(df['Gender'] == 'F') & (df['Department'] == 'CSE') & (df['Marks'] > 80)])
print(df[df['Grade'].isin(['A', 'B'])])
```

```python
import matplotlib.pyplot as plt


df['Department'].value_counts().plot(kind='bar', title='Students per Department')
plt.show()


df['Result'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Result Distribution')
plt.show()


plt.hist(df['Marks'], bins=10, edgecolor='black')
plt.title('Histogram of Marks')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.show()


plt.plot(df['Name'], df['Marks'], marker='o')
plt.title('Marks of Students')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.grid()
plt.show()


fig, axes = plt.subplots(1, 3, figsize=(15, 4))


axes[0].hist(df['Marks'], bins=10, color='skyblue')
axes[0].set_title('Marks Histogram')


df['Grade'].value_counts().plot.pie(ax=axes[1], autopct='%1.1f%%')
axes[1].set_title('Grade Distribution')


df['Department'].value_counts().plot.bar(ax=axes[2])
axes[2].set_title('Dept Count')

plt.tight_layout()
plt.show()
```
